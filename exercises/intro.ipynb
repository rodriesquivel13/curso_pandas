# First examples of Series and Data Frames usings

```python
import pandas as pd
```

```python
ages = [22,13,8,33,25,44,38,22]
```

_Create a serie object from ages list, then print it_

```python
age_serie = pd.Series(ages)
age_serie
```

```python
age_serie.head()
```

***Notice if we don't introduce an argument for head, python shows us the first five records**

_Calculate mean, max and min of `age_serie`_

```python
min__age = age_serie.min()
max_age = age_serie.max()
mean_age = age_serie.mean()
resume_age = [min__age,max_age,mean_age] 
print(f'min, max and average age is {resume_age}, respectivly')
```

***Notice python returns us all three messures like `np.float64 or np.int64` since pandas delagates that job to numpy librarie**
